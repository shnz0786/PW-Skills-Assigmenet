```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """Calculate the probability distribution function (PDF) of a Bernoulli distribution."""
    if x not in [0, 1]:
        return 0
    return p if x == 1 else (1 - p)

p = 0.6
print("Bernoulli PDF (x=0, p=0.6):", bernoulli_pdf(0, p))
print("Bernoulli PDF (x=1, p=0.6):", bernoulli_pdf(1, p))
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """Calculate Z-scores for a given dataset."""
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores


np.random.seed(42)
data = np.random.randn(100)

z_scores = calculate_z_scores(data)

plt.scatter(range(len(z_scores)), z_scores, color='blue', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--', label='Mean (Z=0)')
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.title("Z-scores of the Dataset")
plt.legend()
plt.show()
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
population = np.random.exponential(scale=2, size=10000)
sample_sizes = [5, 10, 30, 50, 100]
means = {size: [] for size in sample_sizes}

for size in sample_sizes:
    for _ in range(1000):
        sample = np.random.choice(population, size=size, replace=True)
        means[size].append(np.mean(sample))

plt.figure(figsize=(12, 6))
for i, size in enumerate(sample_sizes, 1):
    plt.subplot(2, 3, i)
    plt.hist(means[size], bins=30, density=True, alpha=0.75, color='blue', edgecolor='black')
    plt.title(f"Sample Size = {size}")

plt.suptitle("Central Limit Theorem Demonstration (Exponential Distribution)")
plt.tight_layout()
plt.show()
```

```python

#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution .
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, data):
    """Calculate the Z-score of a given data point."""
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (x - mean) / std_dev
    return z_score

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

x_value = 60

z_score = calculate_z_score(x_value, data)
print(f"Z-score for {x_value}: {z_score:.2f}")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)

plt.plot(x, y, color='blue', label='Standard Normal Distribution')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Standard Normal Distribution with Z-score")
plt.legend()
plt.grid()
plt.show()
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-tailed'):
    """Perform hypothesis testing using Z-statistics."""
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = population_std / np.sqrt(sample_size)

    z_score = (sample_mean - population_mean) / standard_error

    if alternative == 'two-tailed':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif alternative == 'left-tailed':
        p_value = stats.norm.cdf(z_score)
    elif alternative == 'right-tailed':
        p_value = 1 - stats.norm.cdf(z_score)
    else:
        raise ValueError("Alternative hypothesis must be 'two-tailed', 'left-tailed', or 'right-tailed'")

    reject_null = p_value < alpha

    return z_score, p_value, reject_null

np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)

population_mean = 50
population_std = 10

z_score, p_value, reject_null = z_test(sample_data, population_mean, population_std)

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")
if reject_null:
    print("Result: Reject the null hypothesis (significant difference)")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference)")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """Calculate the confidence interval for a dataset."""
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)

    z_critical = stats.norm.ppf((1 + confidence) / 2)
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
ci_lower, ci_upper, sample_mean = confidence_interval(sample_data)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print("Interpretation: We are 95% confident that the true population mean falls within this range.")
```

```python
#22. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
normalized_data = zscore(data)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black', density=True)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Original Data Distribution")

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, alpha=0.7, color='red', edgecolor='black', density=True)
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Normalized Data")

plt.tight_layout()
plt.show()


print(f"Original Data: Mean={np.mean(data):.2f}, Std Dev={np.std(data, ddof=1):.2f}")
print(f"Normalized Data: Mean={np.mean(normalized_data):.2f}, Std Dev={np.std(normalized_data, ddof=1):.2f}")
```

```python
#23. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(data):
    """Calculate skewness and kurtosis of a dataset."""
    skewness = skew(data)
    kurt = kurtosis(data)
    return skewness, kurt

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
skewness, kurt = compute_skewness_kurtosis(data)

print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f} (Fisher's definition, 0 indicates normality)")

```
