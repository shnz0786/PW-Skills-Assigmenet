```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Car Sale.csv')
df
```

```python
#1.What is the average selling price of cars for each dealer, and how does it compare across different dealers?

avg_price_by_dealer = df.groupby('Dealer_Name')['Price ($)'].mean().sort_values()

print("Average Selling Price by Dealer ($):")
print(avg_price_by_dealer.round(2))

max_dealer = avg_price_by_dealer.idxmax()
min_dealer = avg_price_by_dealer.idxmin()
price_range = avg_price_by_dealer.max() - avg_price_by_dealer.min()

print(f"\nComparison:")
print(f"Highest average price: {max_dealer} (${avg_price_by_dealer.max():.2f})")
print(f"Lowest average price: {min_dealer} (${avg_price_by_dealer.min():.2f})")
print(f"Price range across dealers: ${price_range:.2f}")
```

```python
#2.Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends?

price_variation = df.groupby('Company')['Price ($)'].std().sort_values(ascending=False)

print("Price Variation (Standard Deviation) by Company ($):")
print(price_variation.round(2))

top_brand = price_variation.idxmax()
top_variation = price_variation.max()

print(f"\nBrand with highest price variation: {top_brand} (${top_variation:.2f})")
```

```python
#3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?
price_stats = df.groupby('Transmission')['Price ($)'].describe()

price_stats['IQR'] = price_stats['75%'] - price_stats['25%']

print("Price Distribution Statistics by Transmission Type:")
print(price_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']].round(2))

print("\nIQR Comparison:")
for transmission in price_stats.index:
    iqr = price_stats.loc[transmission, 'IQR']
    print(f"{transmission} IQR: ${iqr:.2f}")
```

```python
#4.What is the distribution of car prices across different regions?
price_stats = df.groupby('Dealer_Region')['Price ($)'].describe()

price_stats['IQR'] = price_stats['75%'] - price_stats['25%']

print("Price Distribution Statistics by Dealer Region:")
print(price_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']].round(2))
```

```python
#5.What is the distribution of cars based on body styles?
body_style_counts = df['Body Style'].value_counts()
body_style_percent = (body_style_counts / body_style_counts.sum() * 100).round(2)

distribution = pd.DataFrame({
    'Count': body_style_counts,
    'Percentage (%)': body_style_percent
})

print("Distribution of Cars by Body Style:")
print(distribution)
```

```python
#6. How does the average selling price of cars vary by customer gender and annual income?
income_bins = [0, 30000, 60000, 100000, 150000, float('inf')]
income_labels = ['<30K', '30K-60K', '60K-100K', '100K-150K', '>150K']
df['Income Bracket'] = pd.cut(df['Annual Income'], bins=income_bins, labels=income_labels, include_lowest=True)

avg_price = df.groupby(['Gender', 'Income Bracket'])['Price ($)'].mean().unstack().round(2)

avg_price_gender = df.groupby('Gender')['Price ($)'].mean().round(2)

print("Average Selling Price by Gender and Income Bracket ($):")
print(avg_price)
print("\nAverage Selling Price by Gender ($):")
print(avg_price_gender)
```

```python
#7.What is the distribution of car prices by region, and how does the number of cars sold vary by region/
cars_sold = df['Dealer_Region'].value_counts()
cars_sold_percent = (cars_sold / cars_sold.sum() * 100).round(2)

sales_distribution = pd.DataFrame({
    'Number of Cars Sold': cars_sold,
    'Percentage (%)': cars_sold_percent
})

price_stats = df.groupby('Dealer_Region')['Price ($)'].describe()

price_stats['IQR'] = price_stats['75%'] - price_stats['25%']
print("Number of Cars Sold by Dealer Region:")
print(sales_distribution)
print("\nPrice Distribution Statistics by Dealer Region ($):")
print(price_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']].round(2))
```

```python
#8.How does the average car price differ between cars with different engine sizes?
avg_price_engine = df.groupby('Engine')['Price ($)'].mean().sort_values().round(2)

engine_counts = df['Engine'].value_counts()

result = pd.DataFrame({
    'Average Price ($)': avg_price_engine,
    'Number of Cars': engine_counts
})

print("Average Car Price and Count by Engine Type:")
print(result)
```

```python
#9. How do car prices vary based on the customer’s annual income bracket?
income_bins = [0, 30000, 60000, 100000, 150000, float('inf')]

income_labels = ['<30K', '30K-60K', '60K-100K', '100K-150K', '>150K']

df['Income Bracket'] = pd.cut(df['Annual Income'], bins=income_bins, labels=income_labels, include_lowest=True)

price_stats = df.groupby('Income Bracket', observed=True)['Price ($)'].describe()

price_stats['IQR'] = price_stats['75%'] - price_stats['25%']

print("Price Distribution Statistics by Income Bracket ($):")
print(price_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']].round(2))
```

```python
#10.What are the top 5 car models with the highest number of sales, and how does their price distribution look?
top_models = df['Model'].value_counts().head(5)

top_models_df = df[df['Model'].isin(top_models.index)]

price_stats = top_models_df.groupby('Model')['Price ($)'].describe()

price_stats['IQR'] = price_stats['75%'] - price_stats['25%']

print("Top 5 Car Models by Number of Sales:")
print(top_models)
print("\nPrice Distribution Statistics for Top 5 Models ($):")
print(price_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']].round(2))
```

```python
#11..How does car price vary with engine size across different car colors, and which colors have the highest price variation?
avg_price = df.groupby(['Engine', 'Color'])['Price ($)'].mean().unstack().round(2)

price_variation = df.groupby('Color')['Price ($)'].std().sort_values(ascending=False).round(2)

color_counts = df['Color'].value_counts()

variation_result = pd.DataFrame({
    'Price Std Dev ($)': price_variation,
    'Number of Cars': color_counts
})

print("Average Car Price by Engine and Color ($):")
print(avg_price)
print("\nPrice Variation by Color ($):")
print(variation_result)
```

```python
#12.Is there any seasonal trend in car sales based on the date of sale?
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')

df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['Month'].apply(get_season)

monthly_sales = df['Month'].value_counts().sort_index()
monthly_sales.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

seasonal_sales = df['Season'].value_counts().reindex(['Winter', 'Spring', 'Summer', 'Fall'])

monthly_sales_percent = (monthly_sales / monthly_sales.sum() * 100).round(2)
seasonal_sales_percent = (seasonal_sales / seasonal_sales.sum() * 100).round(2)

monthly_distribution = pd.DataFrame({
    'Number of Sales': monthly_sales,
    'Percentage (%)': monthly_sales_percent
})
seasonal_distribution = pd.DataFrame({
    'Number of Sales': seasonal_sales,
    'Percentage (%)': seasonal_sales_percent
})

print("Car Sales Distribution by Month:")
print(monthly_distribution)
print("\nCar Sales Distribution by Season:")
print(seasonal_distribution)
```

```python
#13.How does the car price distribution change when considering different combinations of body style and transmission type?
df['Body_Trans'] = df['Body Style'] + ' - ' + df['Transmission']

price_stats = df.groupby(['Body Style', 'Transmission'])['Price ($)'].describe()

price_stats['IQR'] = price_stats['75%'] - price_stats['25%']

print("Price Distribution Statistics by Body Style and Transmission ($):")
print(price_stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'IQR']].round(2))
```

```python
#14.What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?
print("Unique Engine Values:", df['Engine'].unique())
print("Price ($) Summary:", df['Price ($)'].describe())
print("Annual Income Summary:", df['Annual Income'].describe())
print("Missing Values:\n", df[['Price ($)', 'Engine', 'Annual Income']].isnull().sum())

df['Price ($)'] = pd.to_numeric(df['Price ($)'], errors='coerce')
df['Annual Income'] = pd.to_numeric(df['Annual Income'], errors='coerce')

df['Price ($)'].fillna(df['Price ($)'].median(), inplace=True)
df['Annual Income'].fillna(df['Annual Income'].median(), inplace=True)
df['Engine'].fillna('Unknown', inplace=True)

if df['Engine'].dtype == 'object':

    engine_map = {
        'I4': 1, 'V6': 2, 'V8': 3, 'Electric': 4, 'Unknown': 0
    }
    df['Engine_Encoded'] = df['Engine'].map(engine_map).fillna(0)
else:
    df['Engine_Encoded'] = df['Engine']

print("Engine_Encoded Summary:", df['Engine_Encoded'].describe())

corr_data = df[['Price ($)', 'Engine_Encoded', 'Annual Income']]
corr_matrix = corr_data.corr(method='pearson').round(2)

print("\nCorrelation Matrix (Price, Engine Size, Annual Income):")
print(corr_matrix)

income_bins = [
    df['Annual Income'].min(), 30000, 60000, 100000, 150000, df['Annual Income'].max() + 1
]
income_labels = ['<30K', '30K-60K', '60K-100K', '100K-150K', '>150K']
df['Income_Bracket'] = pd.cut(
    df['Annual Income'], bins=income_bins, labels=income_labels, include_lowest=True
)

price_by_engine_income = df.groupby(
    ['Engine', 'Income_Bracket'], observed=True
)['Price ($)'].mean().unstack().round(2)

print("\nAverage Car Price by Engine and Income Bracket ($):")
print(price_by_engine_income)
```

```python
#15. How does the average car price vary across different car models and engine types?
avg_price = df.groupby(['Model', 'Engine'])['Price ($)'].mean().unstack().round(2)

counts = df.groupby(['Model', 'Engine'])['Price ($)'].count().unstack().fillna(0).astype(int)

print("Average Car Price by Model and Engine ($):")
print(avg_price)
print("\nNumber of Cars by Model and Engine:")
print(counts)
```
